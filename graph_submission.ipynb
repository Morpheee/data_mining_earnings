```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./submission.csv")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x="earnings: 0 no/1 yes", hue="earnings: 0 no/1 yes", stat="proportion")
ratio = df["earnings: 0 no/1 yes"].sum() / len(df)
plt.title("Proportion of labels in predictions. {:.2f} % of True labels.".format(ratio*100))
plt.xticks([0,1])
plt.tight_layout()
plt.savefig("../images/proportion_labels_submission.jpg")
```
